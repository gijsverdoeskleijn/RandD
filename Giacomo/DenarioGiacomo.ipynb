# Checking Environment Variables in a Jupyter Notebook

To check if an environment variable is set in the kernel running your Jupyter notebook, you can use Python's built-in `os` module. This module provides a way to interact with the operating system, including accessing environment variables.

Here's how you can check for environment variables:


```python
import os

# Method 1: Using os.environ dictionary
# This will safely check if a variable exists
def check_env_var(var_name):
    if var_name in os.environ:
        print(f"✅ Environment variable '{var_name}' is set to: {os.environ[var_name]}")
    else:
        print(f"❌ Environment variable '{var_name}' is not set")

# Method 2: Using os.getenv with a default value
# This returns the value or a default if not found
def get_env_var(var_name, default="Not set"):
    value = os.getenv(var_name, default)
    if value == default:
        print(f"❌ Environment variable '{var_name}' is not set")
    else:
        print(f"✅ Environment variable '{var_name}' is set to: {value}")

# Example usage - check for common environment variables
check_env_var("PATH")
check_env_var("HOME")
check_env_var("PYTHONPATH")

# Check a specific variable you're interested in
my_var = "OPENAI_API_KEY"
check_env_var(my_var)

# List all environment variables (be careful, this can be long)
print("\nAll environment variables:")
#for key, value in list(os.environ.items())[:5]:  # Show just first 5 for brevity
for key, value in list(os.environ.items()):
    print(f"{key}: {value}")
print("... and more")
```
